```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 분석
df_final=pd.read_excel('D:\\toyProject_timeManagement\\Time_final_add.xlsx', index_col=None)
df_final.head()
```

```python
#데이터 분석
```

```python
#숫자 열들을 int로 형변환, 첫번째 열 제거
df_final=df_final.drop(['Unnamed: 0'],axis=1)
df_final = df_final.infer_objects()
df_final.info()
df_final.describe()
```

```python
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)


fig = plt.figure(figsize=(20,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df_final['date'],df_final['총합']) ## 선그래프 생성
 
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('총합 그래프',fontsize=20) ## 타이틀 설정
plt.show()
```
