```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
score = pd.read_csv("D:\\김덕성\\영어공부\\토익\\ScoreHistory\\토익점수history.csv")
score
```

```python
sns.pointplot(data=score, x="date", y="Number of correct answers", hue = "part")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.relplot(data=score, x="date", y="-", hue='part', kind="line", col="part", col_wrap=5)
```
